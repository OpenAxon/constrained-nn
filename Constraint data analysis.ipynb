```python

import json
import os
import pandas as pd
```

```python
path_to_results = 'results/fifty-others_five-disjoint'
i = 0
for fi in os.listdir(path_to_results):  # or use 'results'
    with open(os.path.join(path_to_results, fi), 'rt') as f:
        content = json.load(f)
        if i==0:
            columns = content['results'].keys()
            columns.extend(content['config'])
            columns.append('network_type')
            data = pd.DataFrame(columns=columns)
            print data
        data.loc[i] = [None] * len(columns)
    for c in content['results'].keys():
        data.loc[i][c] = content['results'][c]
    for c in content['config'].keys():    
        data.loc[i][c] = content['config'][c]
    data.loc[i]['network_type'] = content['network_type']
    i += 1
    
```

```python
data
```

```python
df1 = data[data['network_type']=='simple']\
[['nb_disjoint_classes', 'nb_other_classes', 'nb_disjoint_constraint_broken', 'nb_other_incorrect']]\
.sort_values(['nb_other_classes'])
```

```python
df2 = data[data['network_type']=='constrained']\
[['nb_disjoint_classes', 'nb_other_classes', 'epochs', 'nb_disjoint_constraint_broken', 'disjoint_classes_output_weight', 'nb_other_incorrect']]\
.sort_values(['nb_other_classes'])
```

```python
pd.merge(df1, df2, on=['nb_disjoint_classes', 'nb_other_classes'], suffixes=['_simple', '_constrained'])
```
